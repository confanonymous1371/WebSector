```python
import json

def load_urls_from_json(json_file_path):
    with open(json_file_path, 'r') as file:
        data = json.load(file)
    urls = [entry['url'] for entry in data.values()]
    return urls

urls = load_urls_from_json("/data/sxs7285/Porjects_code/websector/websector_inference/visualization_data.json")
```

```python
import json

import sys
sys.path.append("/data/sxs7285/Porjects_code/websector/websector_inference")

# Now import the class
from websector_pipeline import WebSectorPipeline,BatchedWebSectorPipeline1

import nest_asyncio
nest_asyncio.apply()

# Initialize your batched pipeline
batched_classifier = BatchedWebSectorPipeline1(
    model_name="Shahriar/WebSector-Flexible", 
    mode="flexible",
    batch_size=4  # Adjust based on your GPU memory
)

# List of URLs to process

# Process URLs in a batch and capture results and timing information
results, timing_info = batched_classifier.process_batch(urls, return_content=False,at_scale = False)




# Print the timing information
print("Timing Information:")
for key, value in timing_info.items():
    print(f"{key}: {value}")

# Save results to a JSON file
with open('batch_results.json', 'w') as f:
    json.dump(results, f, indent=4)

# Save timing info to a separate JSON file
with open('batch_timing_info.json', 'w') as f:
    json.dump(timing_info, f, indent=4)

print("Results saved to 'batch_results.json'")
print("Timing info saved to 'batch_timing_info.json'")


# add hash_key and url to them
```

```python
import pandas as pd
# Convert timing values to minutes where applicable
def convert_to_minutes(timing_dict):
    for key, value in timing_dict.items():
        if 'time' in key:
            timing_dict[key] = value / 60  # Convert seconds to minutes
    return timing_dict

# Convert to minutes
converted_timing_info = convert_to_minutes(timing_info)

# Convert to a Pandas DataFrame for table format
df = pd.DataFrame(list(converted_timing_info.items()), columns=["Process", "Value"])
```
